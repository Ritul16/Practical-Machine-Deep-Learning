# Seperate X & Y

```python
X = df.drop("Class",axis=1)
Y=df['Class']
```

```python
X.shape,Y.shape
```

```python
X.dtypes
```

```python
X.isnull().sum()
```

## Format the target Y

replace M : 1     R : 0

```python
Y[Y=='M']=1
Y[Y=='R']=0
```

```python
Y.unique()
```

```python
Y.dtypes
```

```python
Y = Y.astype("int")
```

```python
Y.dtypes
```

# Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,
                                                 random_state=7,
                                                 stratify=Y)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

# Apply Classification models

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression(penalty='elasticnet',class_weight='balanced',random_state=7,
                        solver='saga',l1_ratio=0.5)
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,Y_pred_lr))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred_lr)
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,Y_train)
```

```python
Y_pred_gnb = gnb.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_gnb))
```

```python
bnb = BernoulliNB()
```

```python
bnb.fit(X_train,Y_train)
```

```python
Y_pred_bnb = bnb.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_bnb))
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(X_train,Y_train)
```

```python
Y_pred_mnb = mnb.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_mnb))
```

```python
svc = SVC(C=1, kernel='rbf', gamma="scale", class_weight='balanced',
          random_state=7)
```

```python
svc.fit(X_train,Y_train)
```

```python
Y_pred_svc = svc.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_svc))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,Y_train)
```

```python
Y_pred_knn = knn.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_knn))
```

```python
dt = DecisionTreeClassifier(random_state=7,class_weight='balanced')
```

```python
dt.fit(X_train,Y_train)
```

```python
Y_pred_dt = dt.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred_dt))
```

# Automate above code

```python
models_list = [
    LogisticRegression(penalty='elasticnet',class_weight='balanced',random_state=7,
                        solver='saga',l1_ratio=0.5),
    GaussianNB(),
    MultinomialNB(),
    SVC(C=1, kernel='rbf', gamma="scale", class_weight='balanced',
          random_state=7),
    KNeighborsClassifier(n_neighbors=3)
]
```

```python
acc_score_list = []
for model in models_list:
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    acc_score_list.append((model.__class__.__name__, accuracy_score(Y_test,Y_pred)))
```
